```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
## Import datasets in python using pandas library
Customers = pd.read_csv('customers.csv')
Products = pd.read_csv('products.csv')
Sales = pd.read_csv('sales.csv')
Stores = pd.read_csv('stores.csv')
```

```python
## Check if their is any null value in the data
print(Customers.isna().sum())
print(Products.isna().sum())
print(Sales.isna().sum())
print(Stores.isna().sum())
```

```python
print(Customers.head())
print(Products.head())
print(Sales.head())
print(Stores.head())
```

```python
## Merge All the tables 

df = Sales.merge(Products , on = 'product_id') \
          .merge(Stores , on = 'store_id') \
          .merge(Customers , on = 'customer_id')

## performing hypothesis testing  

" h0('Null Hypothesis') = 'Both loyal and non-loyal customers have the same average revenue with no significant difference'"
" h1('Alternative Hypothesis') = 'There is a significant difference in average revenue between loyal and non-loyal customers'"
Alpha = 0.05

from scipy.stats import ttest_ind

Loyal_customers = df[df['loyalty_member'] == 1 ]['revenue']
Non_loyal_customers = df[df['loyalty_member'] == 0]['revenue']

t_test , p_value = ttest_ind(Loyal_customers , Non_loyal_customers)

print('T-test statistic : ' , t_test)
print('P-value : ' , p_value)

if p_value < Alpha :
    print('Reject the null hypothesis. There is a significant difference in average revenue between loyal and non-loyal customers.')
else :
    print('Fail to reject the null hypothesis. There is no significant difference in average revenue between loyal and non-loyal customers.')
```

```python
## Hypothesis testing for that does higher discount leads to higher revenue

" H0('Null Hypothesis') = 'There is no significant relationship between discount and revenue' "
" H1('Alternative Hypothesis') = 'There is a significant relationship between discount and revenue' "
Alpha = 0.05
 
from scipy.stats import pearsonr
 
Discount = df['discount']
Revenue = df['revenue']

correlation, p_value = pearsonr(Discount, Revenue)
print('Correlation coefficient : ' , correlation)
print('P-value : ' , p_value)

if p_value < Alpha :
    print('Reject the null hypothesis. There is a significant relationship between discount and revenue.')
else :
    print('Fail to reject the null hypothesis. There is no significant relationship between discount and revenue.')
 
```

```python
## hypothesis testing does higher Weight means higher unit price 

" H0(Null Hypothesis)  = Their is no a Significant Relationship between weight and unit price "
" H1(Alternate Hypothesis) = Their is a  Significant Relationship between weight and  unit Price "

Alpha = 0.05

Weight_g = df['weight_g']
Unit_Price = df['unit_price']

correlation , p_value = pearsonr(Weight_g , Unit_Price)

print('correlation :' , correlation )
print('p_value : ', p_value )

if p_value <= Alpha :
    print('Reject the null hypothesis Their is a Significant relationship between weight and price')
else :
    print('Fail to reject null hypothesis their is no Significant relationship between weight and price' )
```

```python
## Hypothesis testing does higher Cocoa percentage means higher unit price 

" H0 = Their is no significant relationship between cocoa percentagea and unit price"
" H1 = Their is a significant relationship between cocoa percentage and unit price"

Alpha = 0.05

Cocoa_percent = df['cocoa_percent']
Unit_Price = df['unit_price']

correlation , p_value = pearsonr(Cocoa_percent , Unit_Price)

print('Correlation : ', correlation)
print('p_value : ', p_value)

if p_value<= Alpha :
    print('Reject the null hypothesis their is a significant relationship between cocoa percentage and unit price' )
else :
    print('Fail to Reject Null hypothesis their is no significant relationship between cocoa percentage and unit price')
```
